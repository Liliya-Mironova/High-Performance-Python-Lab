# Task 1. Bifurcations

$x_{n+1} = r x_{n}(1-x_{n}) $

$x_0 = \text{rand} (0,1)$

$r = \text{const}$

```python
import numpy as np
import matplotlib.pyplot as plt
import random
%matplotlib inline
```

### 1. Implement the map, plot the evolution of x

```python
def logistic(r, x):
    return r*x*(1 - x)
```

```python
def plot_system(r, x0, n):
    t = np.linspace(0, 1)
    x = x0
    for i in range(n):
        y = logistic(r, x)
        plt.plot([x, x], [x, y], color='tab:blue')
        plt.plot([x, y], [y, y], color='tab:blue')
        x = y

    plt.plot(t, logistic(r, t))
    plt.plot([0, 1], [0, 1])
    plt.grid(True)
    plt.xlabel('$x_{n}$')
    plt.ylabel('$x_{n+1}$')
    plt.title(f"Map")
```

```python
def plot_evolution(r, x0, n):
    x_s = []
    for i in range(n):
        x0 = logistic(r, x0)
        x_s.append(x0)

    plt.plot(range(n), x_s, marker='.')
    plt.grid(True)
    plt.xlabel('$t$')
    plt.ylabel('$N$')
    plt.title("Evolution")

def plot_all(r, x0, n):
    fig = plt.figure(figsize=(14, 6))
    plt.subplot(121)
    plot_system(r, x0, n)

    plt.subplot(122)
    plot_evolution(r, x0, n);
```

```python
plot_all(1., .4, 100) # extinction
```

### Play around with values of r, see the change of evolution

```python
plot_all(2, .4, 100) # certain value
```

```python
plot_all(4, .4, 100) # Jumping
```

```python
plot_all(3, .4, 100) # Jumping
```

### 2. Create a linspace of r’s, for every r save the last “m” values of x after the first “n” values (can be m=200, x=200), play around with values

```python
r_num = 10000
n = 1000
m = 200
x = 200
# x = random.random()

xs = []
r = np.linspace(2.5, 4.0, r_num)
x = 1e-5 * np.ones(r_num)

for i in range(n):
    x = logistic(r, x)
    if i >= n - m:
        xs.append(x)
```

### 3. Plot the bifurcation map

```python
fig = plt.figure(figsize=(12, 8))

for i in range(m):
    plt.plot(r, xs[i], ',k', alpha=.25)

plt.grid(True)
plt.xlabel('Growth Rate - r')
plt.ylabel('Population - N')
plt.title("Bifurcation diagram");
```
